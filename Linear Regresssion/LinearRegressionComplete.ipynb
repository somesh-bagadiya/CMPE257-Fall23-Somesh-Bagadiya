```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import random

## Generates N random points for logistic regression on a line w
def generate_lr(N=20,w0=0,w1=1):
    n = 0
    X1 = []
    Y = []
    while(n < N):
        x1 = random.uniform(-10,10)
        x2 = random.uniform(-1,1)
        y = w0 + w1*x1 + x2
        X1.append(x1)
        Y.append(y)
        n += 1
    data = [np.array([1,X1[i],Y[i]]) for i in range(N)]
    return data
```

```python
f_w0, f_w1 = 1, 1
N = 50
data = generate_lr(N, f_w0, f_w1)
data
```

```python
def abline(slope, intercept, color='b'):
    """Plot a line from slope and intercept"""
    axes = plt.gca()
    x_vals = np.array(axes.get_xlim())
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--', label = "x_2="+str(slope)+"x_1+"+str(intercept), c=color )
```

```python
def lr(X, y, lbda):
  XtX = np.dot(np.transpose(X), X)
  lambda_identity = lbda * np.identity(2)
  w = np.matmul(np.linalg.inv(XtX + lambda_identity) , (np.dot(np.transpose(X),y)))
  return w

plt.scatter(X1,Y, s = 10)
abline(f_w1,f_w0, color="black")

lam = [0.0001, 0.1, 1]

w_lr2 = lr(X, Y, lam[0])
w_lr3 = lr(X, Y, lam[1])
w_lr4 = lr(X, Y, lam[2])

abline(w_lr2[0], w_lr2[1], color='green')
abline(w_lr3[0], w_lr3[1], color='red')
abline(w_lr4[0], w_lr4[1], color='blue')

plt.show()
```

```python
test_i = 0
X_test = []
Y_test = []
X_train = []
Y_train = []
lam = [0.0001, 0.1, 1]
mse = {}

for l in lam:
  for i in range(5):
    test_i = i+1

    X_test = []
    Y_test = []
    X_train = []
    Y_train = []

    X_test = [X[j] for j in range((test_i-1)*10,test_i*10)]
    Y_test = [Y[j] for j in range((test_i-1)*10,test_i*10)]

    for k in range(len(X)):
      if(X[k] not in X_test):
        X_train.append(X[k])
        Y_train.append(Y[k])

    w_lr = lr(X_train, Y_train, l)
    X_corr = [a[1] for a in X_test]
    plt.scatter(X_corr, Y_test, s = 10, color = "blue")
    abline(w_lr[0], w_lr[1], color='green')

    Y_pred = np.dot(X_test, w_lr)
    act_val = np.array(Y_test)
    pred_val = np.array(Y_pred)
    sq_diff = (act_val - pred_val) ** 2
    mse[np.mean(sq_diff)] = w_lr
    # print("Mean Squared Error: ", mse)

  plt.show()


keys = list(mse.keys())
keys.sort()
mse = {i: mse[i] for i in keys}

w_best = list(mse.values())[0]
mse_best = list(mse.keys())[0]
print("Best possible weights are:", w_best, "With Mean Squared error as:", mse_best)
```
